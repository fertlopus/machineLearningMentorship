### Пределы и производные
----

Пределы и производные - область, которая изучает скорость изменения функций. Понимать и вычислять производные важно, так как это основа для огромного количества алгоритмов машинного, а самое главное глубокого обучения в отрасли. Без производных было бы невозможным развитие отрасли и направления как "Искусственный Интеллект". Все строится вокруг них.

Метод оптимизации и нахождения производных скрыт в таких алгоритмах как метод обратного распространения ошибки (backpropagation) и градиентный спуск (плюс различные его варианты).



### Понятие предела функции "на пальцах"

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# создадим линейное пространство точек с которыми будем взаимодействовать
x = np.linspace(-10, 10, 10000) # start, finish, n_points
````

##### Допустим, что **$y = x^2 + 2x + 2$**

````python
y = x**2 + 2*x + 2
````

````python
fig, axes = plt.subplots()
_ = axes.plot(x,y)
````

##### Что мы видим на данном графике функции:

* Отсутсвие линий на графике функции (это нелинейная функция)
* Если мы представим, что мы будем бесконечно близко приближаться к нашей функции, то увидим, что постепенно график функции будет приобретать форму прямой
* Такой подход в теории поможет нам найти касательную $m$ к графику нашей функци

Давайте посмотрим на наши догадки:

````python
fig, axes = plt.subplots()
axes.set_xlim([-1.5, -0.5])
axes.set_ylim([0.5, 1.5])
_ = axes.plot(x,y)
````

Как вы помните из курса университетской высшей математики или может быть кто со школьной программы, пределы функции существуют только для непрерывных функций, в случае если фукнция не непрерывная (имеются точки разрыва или места где функция не определенна).

Основные способы нахождения пределов функций:
* Аналитический - когда мы находим предел функции обычным способом или производя какие-либо алгебраические манипуляции над функцией
* Геометрический метод
* Метод приближений

Возьмем для примера следующий предел функции:


### $\lim _{x \to 1 } \frac{x^2 - 1}{x - 1}$

````python
fig, ax = plt.subplots()

plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

plt.xlim(-5, 10)
plt.ylim(-10, 80)

plt.axvline(x=5, color='purple', linestyle='--')
plt.axhline(y=37, color='purple', linestyle='--')
_ = ax.plot(x,y)
````

Так как находить предел в точке 1 невозможно и функция в данной точке не определенная (это выдаст нам ошибку, что деление на 0 невозможно), то у нас в запасе есть 2 метода:

1) Геометрический - посмотрев на функцию и аппроксимировать предел
2) Метод приближений - подбираясь к значению предела фунции через приближение х

Попробуем реализовать второй способ.

````python
def my_fxn(my_x):
    my_y = (my_x**2 - 1)/(my_x - 1)
    return my_y
````

Аппроксимация слева (-):

````python
points = [2, 0.9, 0.999, 0.9999, 0.99999, 0.999999]
left_limit_points = [my_fxn(point) for point in points]
print(left_limit_points)
````

Аппроксимация справа (+):

````python
points = [1.1, 1.01, 1.001, 1.0001, 1.00001, 1.000001]
right_limit_points = [my_fxn(point) for point in points]
print(right_limit_points)
````

Таким образом можно сделать вывод, что предел функции, стремящийся к 1 приближается к 2:

````python
y = my_fxn(x)

fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axvline(x=1, color='purple', linestyle='--')
plt.axhline(y=2, color='purple', linestyle='--')
_ = ax.plot(x,y)
````

### $\lim _{x \to 0 } \frac{sin(x)}{x}$

##### Самостоятельно проделайте такую же процедуру, плюсом будет если обернете все в функцию которая выводит пределы слева и справа и график :)

### Понятие производной функции "на пальцах"

##### Допустим, что **$y = x^2 + 2x + 2$**

````python
def f(my_x):
    my_y = my_x**2 + 2*my_x + 2
    return my_y
````

````python
y = f(x)
````

````python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
_ = ax.plot(x,y)
````

##### Предлагаю найти касательную к фукнции в точке $x = 2$

````python
# Значение функции в данной точке
f(2)
````

#### Отлично, обзовем данную точку $P$ с координатами в пространстве (2, 10):

````python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10) # scatter plot (точечный график) для визуализации множества либо одной точки
_ = ax.plot(x,y)
````

#### Непосредственный метод поиска производной (**delta method**) через 2 точки в области определения нашей фукнции. Метод заключается в том, чтобы взять 2 точки пространства, которые определены функцией и рассчитать разницу между ними для того чтобы затем вычислить значений касательной к функции. Для целей демонстрации определим 2-ю точку $Q$ с координатами в точке $x=5$


````python
f(5)
````

````python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10)
plt.scatter(5, 37, c = 'orange', zorder=3) # new
_ = ax.plot(x,y)
````

Как находится касательная для координат $P$ и $Q$

<br>

### $m = \frac{\Delta y}{\Delta x} = \frac{y_2 - y_1}{x_2 - x_1} = \frac{37 - 10}{5 - 2} = \frac{27}{3} = 9$

````python
m = (37-10)/(5-2)
m
````

#### Для того, чтобы показать на графике касательную проходящую через координаты точек $P$ и $Q$ нам необходимо решить уравнении прямой на плоскости:

$y = mx + b$

выразим через b
<br>

$b = y - mx$

````python
b = 37-m*5
b
````

````python
line_y = m*x + b
````

````python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10)
plt.scatter(5, 37, c='orange', zorder=3)
plt.ylim(-5, 150) # new
plt.plot(x, line_y, c='orange') # new
_ = ax.plot(x,y)
````

Чем ближе точка $Q$ становится к точке $P$, тем более ближе касательная $m$ прилегает к графику нашей функции. Возьмем для примера точку $Q$ с области x = 2.1

<br>

$\Delta x = x_2 - x_1 = 2.1 - 2 = 0.1$

<br>

Намного меньше чем было в прошлый раз.

````python
f(2.1)
````

````python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10)
plt.scatter(2.1, 10.61, c = 'orange', zorder=3)
_ = ax.plot(x,y)
````

````python
m = (10.61-10)/(2.1-2)
m
````

````python
b = 10.61-m*2.1
b
````

````python
line_y = m*x + b
````

````python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2, 10)
plt.scatter(2.1, 10.61, c='orange', zorder=3)
plt.ylim(-5, 150)
plt.plot(x, line_y, c='orange', zorder=3)
_ = ax.plot(x,y)
````

Чем ближе $Q$ становится к $P$ (\Delta x стремится к 0), тем более точной становится касательная в точке $P = (2, 10)$ которая принимает значение 6.

<br>

Зададим $\Delta x$ бесконечно малое значение например 0.000001

````python
delta_x = 0.000001
delta_x
````

````python
x1 = 2
y1 = 10
````

Выразим $x_2 = x_1 + \Delta x$

````python
x2 = x1 + delta_x
x2
````

Тогда $y_2 = f(x_2)$

````python
y2 = f(x2)
y2
````

Для того чтобы найти касательную к функции $m$  вспомним что у нас есть предел:

<br>

$m = \frac{\Delta y}{\Delta x} = \frac{y_2 - y_1}{x_2 - x_1}$

````python
m = (y2 - y1)/(x2 - x1)
m
````

Таким образом используя данный метод мы нашли касательную к точке $P$ которая равна 6.


**Воспользуйтесь таким подходом, для того чтобы найти касательную в точке х = -1**

````python
x1 = -1
````

````python
y1 = f(x1)
y1
````

````python
# P = (-1, 1)
delta_x
````

````python
x2 = x1 + delta_x
x2
````

````python
b = y2-m*x2
b
````

````python
line_y = m*x + b
````

````python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(x1, y1)
plt.scatter(x2, y2, c='orange', zorder=3)
plt.ylim(-5, 150)
plt.plot(x, line_y, c='orange', zorder=3)
_ = ax.plot(x,y)
````

### Производная параметрической фукнции

Определим функцию $f(x,y)$ для $z = x^2 - y^2$

````python
def f(my_x, my_y):
    return my_x**2 - my_y**2
````

Визуализируем график области определения функции относительно х принимая во внимание что у константное значение

````python
xs = np.linspace(-3, 3, 1000)
````

````python
zs_wrt_x = f(xs, 0)
````

````python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

plt.xlabel('x')
plt.ylabel('z', rotation=0)
_ = ax.plot(xs, zs_wrt_x)
````

Чтобы определить наклон к заданной точке кривой мы можем использовать частную производную  $\frac{\delta z}{\delta x} = 2x$

````python
def delz_delx(my_x, my_y): # y isn't relevant
    return 2*my_x
````

````python
x_samples = [-2, -1, 0, 1, 2]
colors = ['red', 'orange', 'green', 'blue', 'purple']
````

````python
def point_and_tangent_wrt_x(my_xs, my_x, my_y, my_f, fprime, col):

    my_z = my_f(my_x, my_y) # z = f(x, y)
    plt.scatter(my_x, my_z, c=col, zorder=3)

    tangent_m = fprime(my_x, my_y) # Slope is partial derivative of f(x, y) w.r.t. x
    tangent_b = my_z - tangent_m*my_x # Line is z=mx+b, so b=z-mx
    tangent_line = tangent_m*my_xs + tangent_b

    plt.plot(my_xs, tangent_line, c=col,
             linestyle='dashed', linewidth=0.7, zorder=3)
````

````python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

for i, x in enumerate(x_samples):
    point_and_tangent_wrt_x(xs, x, 0, f, delz_delx, colors[i])

plt.ylim(-1, 9)
plt.xlabel('x')
plt.ylabel('z', rotation=0)
_ = ax.plot(xs, zs_wrt_x)
````

````python
def delz_dely(my_x, my_y):
    return -2*my_y
````

````python
y_samples = [-2, -1, 0, 1, 2]
````

````python
def point_and_tangent_wrt_y(my_ys, my_x, my_y, my_f, fprime, col): # changed my_xs to my_ys

    my_z = my_f(my_x, my_y)
    plt.scatter(my_y, my_z, c=col, zorder=3) # changed my_x to my_y

    tangent_m = fprime(my_x, my_y)
    tangent_b = my_z - tangent_m*my_y # changed my_x to my_y
    tangent_line = tangent_m*my_ys + tangent_b # changed my_xs to my_ys

    plt.plot(my_ys, tangent_line, c=col,
             linestyle='dashed', linewidth=0.7, zorder=3) # changed my_xs to my_ys
````

````python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

for i, y in enumerate(y_samples):
    point_and_tangent_wrt_y(ys, 2, y, f, delz_dely, colors[i])

plt.ylim(-5, 5)
plt.xlabel('y')
plt.ylabel('z', rotation=0)
_ = ax.plot(xs, zs_wrt_y)
````
